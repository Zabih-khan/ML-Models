```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("Dummy Variables & One Hot Encoding/homeprices (2).csv")
```

```python
df.head()
```

```python
encoder = LabelEncoder()

labelencoder = encoder.fit_transform(df.town)
```

```python
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(

    transformers= [("one_hot_encoder", OneHotEncoder(), [0])],
    remainder="passthrough"

)
```

```python
ct.fit_transform(df)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(ct.fit_transform(df), df.price)
```

```python
model.predict(ct.fit_transform(df))
```
